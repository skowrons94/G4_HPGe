```python
import ROOT

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

from glob import glob
from tqdm import tqdm
```

```python
a = 3.78288e-03
b = 6.57475e-04
c = 1.54547e-08

MevToKev = 1e3

def get_reso( energy ):
    return np.sqrt( a + b * energy + c * energy**2 )
```

```python
for file in glob("data/20_cm/*.root"):

    name = file.split( "/" )[2][:-5]
    events = { "Energy" : [] }

    try:
        f1 = ROOT.TFile( file )
    except:
        print( "Error opening file: {}".format( file ) )
        continue

    t1 = f1.Get( "t1" )

    for idx in tqdm( range( t1.GetEntries( ) ) ):

        t1.GetEntry( idx )

        energy = getattr( t1, "Energy" ) * MevToKev

        if( energy > 0 ): energy = np.random.normal( energy, get_reso( energy ) )

        events["Energy"].append( energy )

    df = pd.DataFrame( events )
    df.to_pickle( "processed/20_cm/{}.pkl".format(name) )
```
